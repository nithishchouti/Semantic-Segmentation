```python
import torch
import time
from MobilenetV2 import SegmentationCustom as v2
from MobilenetV3 import SegmentationCustom as v3
```

```python
model_v2 = v2(num_classes=21, output_stride=16)
model_v3 = v3(num_classes=21, output_stride=16)
```

```python
model_v2.to("cuda")
model_v2.eval()
```

```python
model_v3.to("cuda")
model_v3.eval()
```

```python
tensor = torch.randn(1,3,512,512).to("cuda")
```

```python
time_taken = []
for i in range(40):
    start = time.time()
    with torch.no_grad():
        _ = model_v2(tensor)
    end = time.time()
    print("Time taken: ", end-start)
    if i > 5:
      time_taken.append(end-start)
    
print("Average across 40 frames: ", sum(time_taken) / len(time_taken))

```

```python
time_taken = []
for _ in range(40):
    start = time.time()
    with torch.no_grad():
        _ = model_v3(tensor)
    end = time.time()
    print("Time Taken: ", end-start)
    if i > 5:
      time_taken.append(end-start)
    
print("Average across 40 frames: ", sum(time_taken) / len(time_taken))

```
